# Releases

```python
child_categories = pd.DataFrame(
  requests.get(
      BASE_URL + "/release/series?release_id=80&file_type=json&api_key=" + API_KEY
  ).json()['seriess']
)

child_categories.head()
```

# Series

```python
query_series = pd.DataFrame(
  requests.get(
      BASE_URL + "/series/search?search_text=interest+rates&file_type=json&api_key=" + API_KEY
  ).json()['seriess']
)

query_series.head()
```

# Sources

```python
inspect_source = requests.get(
      BASE_URL + "/source?source_id=31&file_type=json&api_key=" + API_KEY
  ).json()

inspect_source
```
